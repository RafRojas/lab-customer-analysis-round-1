```python
#1 Read the 3 files into python as data frames.

import pandas as pd

file1 = pd.read_excel('file1.xlsx')
file2 = pd.read_excel('file2.xlsx')
file3 = pd.read_excel('file3.xlsx')

#2 Dataframe's shape
print('The shape of the file one is ', str(file1.shape))
print('The shape of the file two is ', str(file2.shape))
print('The shape of the file three is ', str(file3.shape))
```

```python
#3 Standardize header names.

file1.columns = [x.lower() for x in file1.columns]
file2.columns = [x.lower() for x in file2.columns]
file3.columns = [x.lower() for x in file3.columns]

display(file1)
 
```

```python
# 5 Concatenate the three dataframes

df = pd.concat ([file1, file2, file3], axis = 0)

display(df)

# 6 Which columns are numerical? The numerical columns are: 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints' and 'total claim amount'

# 7 Which columns are categorical? 'customer', 'st', 'education', 'policy type', 'vehicle class' and 'state'
```

```python
# 8) Perform the data cleaning operations mentioned so far in class

# a) Delete the column education and the number of open complaints from the dataframe.

df1 = df.drop('education', axis = 1)
df1 = df.drop('number of open complaints', axis = 1)

# total_rows['ColumnID'] = total_rows['ColumnID'].astype(str)

df1['customer lifetime value'] = df1['customer lifetime value'].astype(str)

df1['customer lifetime value'] = df1['customer lifetime value'].str.replace('%', '')

display(df1)
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

df1['customer lifetime value'] =  pd.to_numeric(df1['customer lifetime value'], errors='coerce') * 100

display(df1)
```

```python
# Check for duplicate rows in the data and remove if any.

df1.loc[df1.duplicated(keep = False), :]

df2 = df1.drop_duplicates()
```

```python
# Filter out the data for customers who have an income of 0 or less.

df_low_income = df2[df2['income'] <= 0]

display(df_low_income)
```
